```python
import rarfile
import os

rar_path = "Fingernails.rar"  # Change this if needed
extract_path = "Fingernails_extracted"

# Ensure rarfile is enabled
rarfile.UNRAR_TOOL = "C:/Program Files/Unrar/unrar.exe"  # Set correct path to UnRAR if needed

# Extract the archive
with rarfile.RarFile(rar_path, 'r') as rf:
    rf.extractall(extract_path)

# Check extracted folders
os.listdir(extract_path)
```

```python
for folder in os.listdir(extract_path):
    print(folder, "->", len(os.listdir(os.path.join(extract_path, folder))), "images")
```
